<b>Word Embedding Techniques using Embedding layer in keras.</b>

```python
## Libraries used Tensorflow 2.0 and keras
from keras.preprocessing.text import one_hot
```

```python
## Libraries used Tensorflow > 2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences
sents = ['the glass of milk', 
        'the glass of juice', 
        'the cup of tea', 
        'I am a good devloper',
        'understand the meaning of words',
        'your videos are good']
```

```python
sents
```

```python
# Vocabulary size
voc_size = 10000
```

<b>One Hot Representation</b>

```python
onehot_repr = [one_hot(sent,voc_size) for sent in sents]
onehot_repr
```

<b>Word Embedding Representation</b>

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_len = 8
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_len)
embedded_docs 
```

```python
dim = 15
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
